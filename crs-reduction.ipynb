```python
%display latex
R.<alpha,beta,gamma,delta,x> = PolynomialRing(SR)
n = 5
m = 4
l = m - 2

def defpoly(label, d): 
    poly_coeffs = list(var(label+'_%d' % i) for i in range(d+1))
    poly = sum(c*x**i for i,c in enumerate(poly_coeffs))  
    return (poly,poly_coeffs)
    
(t,t_coeffs) = defpoly('t',n)

us = []
uscoeffs = []
vs = []
vscoeffs = []
ws = []
wscoeffs = []

for i in range(m):
    (u,u_coeffs) = defpoly('u_%d'%i,n-1)
    us.append(u); uscoeffs.append(u_coeffs)
    
    (v,v_coeffs) = defpoly('v_%d'%i,n-1)
    vs.append(v); vscoeffs.append(v_coeffs)
    
    (w,w_coeffs) = defpoly('w_%d'%i,n-1)
    ws.append(w); wscoeffs.append(w_coeffs)

allcoeffs = flatten(uscoeffs)+flatten(vscoeffs)+flatten(wscoeffs)+t_coeffs
    
print(t,us,vs,ws)
show(allcoeffs)
```

```python
# Gro16 CRS
R1 = [alpha,gamma,x,alpha*delta,gamma*delta*t,gamma**2*delta*t**2,(alpha+beta)*gamma*delta*t] + \
        list(gamma*delta*x**i for i in range(n)) + \
        list(gamma**2*delta*t*x**i for i in range(n)) + \
        list(gamma*delta*ws[i] + delta*(alpha+beta)*u[i] for i in range(l+1)) + \
        list(gamma**2*delta*w[i]+gamma*delta*(alpha+beta)*u[i] for i in range(l+1,m+1))

show(R1)
print(len(R1))

R2 = [1, beta, delta, beta*delta, gamma*delta*t, delta**2] + \
        list(gamma*delta*x**i for i in range(n))

show(R2)
print(len(R2))
```

```python
A_coeffs = list(var('A_%d' % i) for i in range(len(R1)))
A = sum(c*x for c,x in zip(A_coeffs,R1))

B_coeffs = list(var('B_%d' % i) for i in range(len(R2)))
B = sum(c*x for c,x in zip(B_coeffs,R2))

C_coeffs = list(var('C_%d' % i) for i in range(len(R1)))
C = sum(c*x for c,x in zip(C_coeffs,R1))

show(A)
show(B)
```
